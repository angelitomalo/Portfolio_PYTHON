```python
#Proyecto Python para analizar el cambio promedio de cambios en el nivel del mar en el mundo desde 1880 y predecir el cambio hasta 2050

#Crear un scatter plot que muestre el año y el nivel del mar ajustado
#Usar una línea de regresión para predecir el nivel del mar hasta 2050
#Usar una nueva línea de regresión con los datos únicamente desde el año 2000 al año más reciente del dataset y que llegue hasta 2050
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
df=pd.read_csv(r"C:\Users\PC\Downloads\epa-sea-level.csv")
```

```python
#Scatter plot
fig,ax=plt.subplots(figsize=(15,5))
ax.scatter(df["Year"], df["CSIRO Adjusted Sea Level"], label='Datos originales')

#Crear línea de regresión
regress=linregress(df["Year"],df["CSIRO Adjusted Sea Level"])
years_future=range(1880,2051)
ax.plot(years_future, regress.intercept + regress.slope*years_future, 'r', label='Predicción total')

#Crear segunda línea de regresión
df_2000=df.loc[df["Year"]>=2000]
regress_2000=linregress(df_2000["Year"],df_2000["CSIRO Adjusted Sea Level"])
ax.plot(years_future, regress_2000.intercept + regress_2000.slope*years_future, 'g', label='Predicción desde el año 2000')
plt.legend()

ax.set_xlabel("Año")
ax.set_ylabel("Nivel del mar en pulgadas")
ax.set_title("Aumento del nivel del mar")
```
